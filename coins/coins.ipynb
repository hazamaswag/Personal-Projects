```python
# probability of getting at least a single head in various n coin tosses.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def graph_me(n):
    iters = range(1,1000)
    coins = 0
    empirical = []
    for i in iters:
        if sum(np.random.randint(2, size=n)) > 0:
            coins += 1
        empirical += [coins / i]
    plt.plot(iters, empirical, label=n)
    plt.legend()
```

```python
# empirical result: 
for i in range(1,6):
    graph_me(i)
```

```python
# actual answer is:
num_coins = np.arange(10)
actual = 1 - (1/2**num_coins)
plt.plot(num_coins, actual)
```

```python
# final thoughts:
# As, the number of coins increases, the chances to get a single head in coin tosses is much higher, this part is trivial.
# What's cool is this is actually a decreasing graph, not linear or exponential. So, the more the coins, the closer the
# probability is to get to 1, but this convergence is a decreasing growth, which is very interesting.
```
